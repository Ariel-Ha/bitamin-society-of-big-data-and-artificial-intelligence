### [1-1] 
datetime 열을 년-월-일, 시간 이렇게 분리해서 새로운 tempDate 라는 이름의 열을 만들어주세요.
* ex) 2011-01-01 00:00:00  ->  2011-01-01 , 00:00:00	

####[1-1] sol 1)

```python
#datetime 열을, 년-월-일, 시간 분리
train['tempDate']= train['datetime'].str.split(' ')
train['tempDate']
```

####[1-1] sol 2)

```python
train['tempDate']=train.datetime.apply(lambda x:x.split())
train['tempDate']
```

### [1-2] 
1-1에서 분리한 tempDate열을 가지고, train 데이터에 year, month, day, weekday 열을 추가해주세요.

####[1-2] sol 1)

```python
#sol1) 분리한 tempDate를 가지고 year, month, day, weekday column 추가
train['year']= train['tempDate'].str[0].str[:4] #tempDate의 첫 열의 연도만 출력
train['month']= train['tempDate'].str[0].str[5:7]
train['day']= train['tempDate'].str[0].str[8:10]

# train['weekday']= train['tempDate'].str[0].weekday() -> 아래에서 진행

hour = train['tempDate'].str[1].astype(str).str.split(':')
train['hour']= hour.str.get(0)
train['hour']
```

weekday 함수를 사용하기 위해서는 날짜 컬럼의 데이터 타입을 datetime 으로 바꿔줘야 한다.<br>
현재 tempDate의 첫 열인 2000-00-00 형태의 날짜 데이터는 object 타입이다. 

```python
#tempDate 열의 datatype을 datetime으로 바꿔주기
train['tempDate'] = pd.to_datetime(train['tempDate'].str[0], format='%Y-%m-%d', errors='raise')
```

```python
#데이터 타입 변경 결과 확인
train.info()
```

```python
# weekday 함수는 날짜의 요일 정보를 숫자값으로 리턴해준다. 0부터 6까지 리턴하며, 0 월요일~6 일요일
#train 데이터 프레임의 weekday열 추가
train['weekday'] = train['tempDate'].dt.weekday
```

#### [1-2] sol 2)

```python
#분리한 tempDate를 가지고 year, month, day, weekday column 추가
train['year']=train.tempDate.apply(lambda x:x[0].split('-')[0])
#tempDate 데이터 프레임을 매개변수 x로 받고 x의 0번째 인덱스를 -를 기준으로 분리한다. 분리된 것에서 0번째 인덱스를 train의 year 데이터로 저장
train['month']=train.tempDate.apply(lambda x:x[0].split('-')[1])
train['day']=train.tempDate.apply(lambda x:x[0].split('-')[2])
train['weekday']=train.tempDate.apply(lambda x:calendar.day_name[datetime.strptime(x[0], "%Y-%m-%d").weekday()])
'''tempDate 데이터 프레임를 매개변수 x로 받고 calendar의 day_name 요일을 표현하는 함수를 사용하는데 인자로 다음을 받는다.
[datetime의 strptime 메소드를 통해 tempDate 의 첫 번째 열인 숫자로 보이지만 문자형인 날짜에 대해 datetime 타입으로 변경한다. 
이후 요일을 숫자로 변환하는 weekday 함수를 사용한다. 숫자로 변환된 요일은 다시 day_name을 통해 Saturday와 같이 출력된다.'''
train['hour']=train.tempDate.apply(lambda x:x[1].split(":")[0])
```

### [1-3] 
1-2에서 만든 year, month, day, hour의 문자형 데이터를 숫자형 데이터로 변환해주세요.

#### [1-3] sol 1)

```python
#분리한 형태의 문자형 데이터를 숫자형 데이터로 변환
#다수의 열 변환
train[['year', 'month', 'day', 'hour']] = train[['year', 'month', 'day', 'hour']].apply(pd.to_numeric)
train.dtypes
```

#### [1-3] sol 2)

```python
#분리한 형태의 문자형 데이터를 숫자형 데이터로 변환
train['year']=pd.to_numeric(train.year, errors="coerce") #errors=coerce의 의미 : 만약 숫자로 변경할 수 없는 데이터라면 기존 데이터를 지우고 NaN으로 설정하여 반환
train['month']=pd.to_numeric(train.month, errors="coerce")
train['day']=pd.to_numeric(train.day, errors="coerce")
train['hour']=pd.to_numeric(train.hour, errors="coerce")
```

### [1-4] 
이제, 필요없는 tempDate 열을 삭제해 주시고, train.csv을 위에서부터 5행 출력해주세요.

```python
#필요없는 tempDate열 삭제 후, train을 위에서부터 5행 출력
train.drop(['tempDate'], axis=1, inplace=True) #inplace=True로 자동으로 train 업데이트
train.head()
```
